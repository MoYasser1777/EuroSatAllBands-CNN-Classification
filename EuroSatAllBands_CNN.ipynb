## Convolutional Neural Networks (CNN)

Machine learning is a field of artificial intelligence that enables systems to learn from data and make decisions without being explicitly programmed. It involves algorithms that iteratively learn from data and improve their accuracy over time by optimizing an error metric, typically through the use of loss functions. These functions quantify the difference between the predicted outputs and the actual outputs, guiding the algorithm to minimize this error during training. By continually adjusting and improving, machine learning models can achieve remarkable accuracy in tasks ranging from simple classification to complex decision-making scenarios.

Convolutional Neural Networks (CNNs) are a specialized kind of neural networks particularly effective for analyzing visual imagery. They employ a mathematical operation known as convolution, which uses sliding window techniques to process data. This method is highly efficient in recognizing patterns and features in images due to the sparse interactions and parameter sharing of convolutional layers. These characteristics allow CNNs to capture local patterns like edges and textures with a significantly reduced amount of parameters compared to fully connected networks, making them highly efficient for tasks involving high-dimensional data like images.

### Key Components of CNN Architecture

The architecture of a Convolutional Neural Network (CNN) typically includes several key components, each playing a crucial role in the network's ability to process and interpret visual information:

- **Convolutional Layers**:
  - Apply a number of filters to the input.
  - Create feature maps that capture essential details within the data.

- **Pooling Layers**:
  - Often follow convolutional layers.
  - Reduce the spatial size of the feature maps, decreasing the computational load.
  - Extract dominant features that are invariant to small changes in the input.

- **Fully Connected Layers**:
  - Interpret the features extracted by convolutional and pooling layers.
  - Perform classification or regression tasks based on the interpreted features.

- **Softmax Layer**:
  - Typically used in the final layer if the task is classification.
  - Outputs the probabilities of the instance belonging to each class.

- **Dropout Layers**:
  - Included to prevent overfitting.
  - Randomly drop units from the neural network during the training process to improve generalization.

### Overview of Neural Network Hyperparameters

In training neural networks, various hyperparameters must be predefined to guide the training process. These include:

- **Batch Size**
  - **Definition**: Number of training examples used per iteration.
  - **Common Values**: 32, 64, 128, 256.

- **Learning Rate**
  - **Definition**: Step size at each iteration to minimize the loss function.
  - **Common Values**: Often starts small, e.g., 0.01, 0.001, and can be dynamically adjusted (lowered as we get nearer to minima, check scheduers).

- **Optimizer Type**
  - **Examples**: SGD (Stochastic Gradient Descent), Adam, RMSprop, Adagrad.
  - **Role**: Different optimizers affect training dynamics and model performance differently.

- **Beta Parameters for Adam Optimizer**
  - **Beta1 and Beta2**: Control the decay rates of moving averages of past gradients and squared gradients.
  - **Common Values**: Beta1 = 0.9, Beta2 = 0.999.

- **Epochs**
  - **Definition**: One full pass of the training dataset through the learning algorithm.
  - **Usage**: More epochs can lead to better learning but risk overfitting.

- **Momentum**
  - **Definition**: Helps accelerate SGD and dampens oscillations.
  - **Common Values**: 0.9, 0.99.

- **Regularization Parameters (L1 & L2)**
  - **Purpose**: Prevent overfitting by adding a penalty on the size of the coefficients.
  - **Parameters**: Lambda or alpha values for L1 (Lasso) and L2 (Ridge) regularization.

- **Dropout Rate**
  - **Definition**: Fraction of neurons to randomly drop during training.
  - **Common Values**: 0.2, 0.3, 0.5.

- **Number of Layers and Number of Neurons in Each Layer**
  - **Role**: Determines the architecture's depth and width, affecting its ability to capture complex patterns.

- **Error Function Type for Classification with Softmax**
  - **Definition**: Cross-entropy loss, also known as log loss.
  - **Purpose**: Measures the performance of a classification model whose output is a probability value between 0 and 1. Cross-entropy loss increases as the predicted probability diverges from the actual label (an alternative can be Categorical Hinge Loss).

- **Early Stopping**
  - **Purpose**: Halt training when performance on a validation set starts to worsen to prevent overfitting.

Each of these hyperparameters can significantly impact the effectiveness and efficiency of the neural network training process, and tuning them appropriately is crucial for achieving optimal performance.

````python
import os
import numpy as np
import matplotlib.pyplot as plt
from skimage import io,transform
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
import torch.optim as optim
from tqdm import tqdm
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score,ConfusionMatrixDisplay

import cv2
````

````python
# Define dataset path and classes
DATASET_PATH = "EuroSATallBands/ds/images/remote_sensing/otherDatasets/sentinel_2/tif"
MODEL_PATH = "cnn_model.pth"
CLASS_NAMES = ["Residential", "River", "Forest"]
CLASS_NUMBERS = ["0","1","2"]

BATCH_SIZE = 64
EPOCHS_NUM = 20
ALPHA = 0.001
MOMENTUM = 0.9

CONV1_INPUT = 13
CONV1_OUTPUT = 6

CONV2_INPUT = 6
CONV2_OUTPUT = 16

CONV_KERNAL_SIZE = 5
POOL_KERNAL_SIZE = 2

CONV_STRIDE = 1
POOL_STRIDE = 2

FC1_NEURONS = 120
FC2_NEURONS = 84

CLASSES_NUM = 3

LENET_5_INPUT_DIM = 32
BANDS_NUM = 13

TORCH_SEED = 47
````

````python
import torch
import torchvision
print(torch.__version__)
print(torchvision.__version__)
````

````python
# DO NOT CHANGE THIS CELL
## Training set indices.
np.random.seed(27)  # Set random seed for reproducibility

# Randomly select indices for the test sets for each class
residential_indices = np.random.choice(np.arange(3000), size=600, replace=False)
forest_indices = np.random.choice(np.arange(3000), size=600, replace=False)
river_indices = np.random.choice(np.arange(2500), size=500, replace=False)

residential_val_indices = residential_indices[:300]
forest_val_indices = forest_indices[:300]
river_val_indices = river_indices[:250]

residential_test_indices = residential_indices[300:]
forest_test_indices = forest_indices[300:]
river_test_indices = river_indices[250:]
````

````python
# Dictionary to store images with their classes
images_classes = {}

residential_labels = []
river_labels = []
forest_labels = []

for class_name in CLASS_NAMES:
    class_dir = os.path.join(DATASET_PATH, class_name)
    
    if os.path.exists(class_dir):
        images_classes[class_name] = []  
        
        for img_filename in os.listdir(class_dir):
            img_path = os.path.join(class_dir, img_filename)
            
            image = io.imread(img_path)

            image = image.astype(np.float32)

            image = (image - image.min()) / (image.max() - image.min())
            
            # Resize image to (32, 32, 13) - Keeping 13 bands for multispectral data
            image_resized = transform.resize(image, (LENET_5_INPUT_DIM, LENET_5_INPUT_DIM, BANDS_NUM))
            
            images_classes[class_name].append(image_resized)
            
            if class_name == "Residential":
                residential_labels.append("Residential")
            elif class_name == "River":
                river_labels.append("River")
            elif class_name == "Forest":
                forest_labels.append("Forest")

# Convert image lists to NumPy arrays
residential_images = np.array(images_classes.get("Residential", []))
river_images = np.array(images_classes.get("River", []))
forest_images = np.array(images_classes.get("Forest", []))

# Convert labels lists to NumPy arrays
residential_labels = np.array(residential_labels)
river_labels = np.array(river_labels)
forest_labels = np.array(forest_labels)

# verify data loading
print("Residential images shape:", residential_images.shape)
print("River images shape:", river_images.shape)
print("Forest images shape:", forest_images.shape)
````

````python
# Take the rest of data from val/test to be the training data
residential_training_indices = np.setdiff1d(np.arange(3000), residential_indices)
forest_training_indices = np.setdiff1d(np.arange(3000), forest_indices)
river_training_indices = np.setdiff1d(np.arange(2500), river_indices)

# Split dataset and labels into training, validation, and test sets
train_residential = residential_images[residential_training_indices]
train_residential_labels = residential_labels[residential_training_indices]

train_river = river_images[river_training_indices]
train_river_labels = river_labels[river_training_indices]

train_forest = forest_images[forest_training_indices]
train_forest_labels = forest_labels[forest_training_indices]

test_residential = residential_images[residential_test_indices]
test_residential_labels = residential_labels[residential_test_indices]

test_river = river_images[river_test_indices]
test_river_labels = river_labels[river_test_indices]

test_forest = forest_images[forest_test_indices]
test_forest_labels = forest_labels[forest_test_indices]

val_residential = residential_images[residential_val_indices]
val_residential_labels = residential_labels[residential_val_indices]

val_river = river_images[river_val_indices]
val_river_labels = river_labels[river_val_indices]

val_forest = forest_images[forest_val_indices]
val_forest_labels = forest_labels[forest_val_indices]

# Concatenate training, validation, and test data
train_data = np.concatenate((train_residential, train_river, train_forest), axis=0)
train_labels = np.concatenate((train_residential_labels, train_river_labels, train_forest_labels), axis=0)

test_data = np.concatenate((test_residential, test_river, test_forest), axis=0)
test_labels = np.concatenate((test_residential_labels, test_river_labels, test_forest_labels), axis=0)

val_data = np.concatenate((val_residential, val_river, val_forest), axis=0)
val_labels = np.concatenate((val_residential_labels, val_river_labels, val_forest_labels), axis=0)

# Encode labels using LabelEncoder ("Residential", "River", "Forest") --> ("0", "1", "2")
label_encoder = LabelEncoder()
train_labels = label_encoder.fit_transform(train_labels)
test_labels = label_encoder.transform(test_labels)
val_labels = label_encoder.transform(val_labels)

#verify loading and preprocessing
print("Train Data Shape:", train_data.shape)
print("Train Labels Shape:", train_labels.shape)
print("Test Data Shape:", test_data.shape)
print("Test Labels Shape:", test_labels.shape)
print("Validation Data Shape:", val_data.shape)
print("Validation Labels Shape:", val_labels.shape)
````

````python
import torch
print(torch.__version__)  # Check PyTorch version
print(torch.version.cuda)  # Check CUDA version used by PyTorch
print(torch.cuda.is_available())  # Check if CUDA is detected
print(torch.cuda.device_count())  # Number of available GPUs
print(torch.cuda.get_device_name(0) if torch.cuda.is_available() else "No GPU detected")
````

````python
# Convert data to tensors
train_data_tensor = torch.tensor(train_data).permute(0, 3, 1, 2)  # (H, W, C) -> (C, H, W)
train_labels_tensor = torch.tensor(train_labels)

test_data_tensor = torch.tensor(test_data).permute(0, 3, 1, 2)
test_labels_tensor = torch.tensor(test_labels)

val_data_tensor = torch.tensor(val_data).permute(0, 3, 1, 2)
val_labels_tensor = torch.tensor(val_labels)  

# Create datasets
train_dataset = TensorDataset(train_data_tensor, train_labels_tensor)
test_dataset = TensorDataset(test_data_tensor, test_labels_tensor)
val_dataset = TensorDataset(val_data_tensor, val_labels_tensor)

torch.manual_seed(TORCH_SEED)

# Create data loaders
train_dataloader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)
val_dataloader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)
````

````python
# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
````

````python
"""
    LeNet-5 was chosen as the base architecture because it is a simple yet effective CNN designed for image classification tasks.
    It provides a good balance between computational efficiency and feature extraction,
    making it suitable for satellite image classification on EuroSAT.

    This architecture is based on LeNet-5 but modified for EuroSAT's 13-channel satellite images.
    The key changes include increasing the input channels from 1 to 13 to process multi-spectral data,
    adjusting convolutional layers with custom filter sizes and strides to better extract spatial features,
    and modifying the fully connected layers to classify three categories (residential, river, forest) instead of LeNet-5's 10-digit output. 

    We don't add softmax layer because PyTorch's CrossEntropyLoss internally applies log_softmax to the model's raw outputs (logits).
"""

def create_model():
    model = nn.Sequential(
        nn.Conv2d(CONV1_INPUT, CONV1_OUTPUT, kernel_size=CONV_KERNAL_SIZE,stride=CONV_STRIDE),
        nn.ReLU(),
        nn.MaxPool2d(kernel_size=POOL_KERNAL_SIZE, stride=POOL_STRIDE),
        
        nn.Conv2d(CONV2_INPUT, CONV2_OUTPUT, kernel_size=CONV_KERNAL_SIZE,stride=CONV_STRIDE),
        nn.ReLU(),
        nn.MaxPool2d(kernel_size=POOL_KERNAL_SIZE, stride=POOL_STRIDE),
        
        nn.Flatten(),
        nn.Linear(CONV2_OUTPUT * CONV_KERNAL_SIZE**2, FC1_NEURONS),
        nn.ReLU(),
        nn.Linear(FC1_NEURONS, FC2_NEURONS),
        nn.ReLU(),
        nn.Linear(FC2_NEURONS, CLASSES_NUM)  
    )
    return model.to(device)
````

````python
# Initialize loss function
criterion = nn.CrossEntropyLoss()

def train_model(model, train_dataloader, num_epochs, lr, momentum):
    optimizer = optim.SGD(model.parameters(), lr=lr, momentum=momentum)
    
    for epoch in range(num_epochs):
        running_loss = 0.0
        for images, labels in tqdm(train_dataloader, desc=f"Epoch {epoch+1}/{num_epochs}"):
            images = images.float().to(device)
            labels = labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        avg_loss = running_loss / len(train_dataloader)
        #print(f"Epoch {epoch+1}/{num_epochs}, Loss: {avg_loss:.4f}")

    print("Training is Finished")

    # Save the trained model
    torch.save(model.state_dict(), MODEL_PATH)
    print(f"Model saved at {MODEL_PATH}")

def evaluate_model(model, val_dataloader, load_path, class_names):
    # Load the model
    model.load_state_dict(torch.load(load_path))
    model.to(device)
    model.eval()  # Set model to evaluation mode

    correct_predictions = 0
    total_predictions = 0
    pred_labels = []
    true_labels = []
    misclassified_samples = []

    with torch.no_grad():
        for images, labels in tqdm(val_dataloader, desc="Evaluation"):
            images = images.float().to(device)
            labels = labels.to(device)

            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total_predictions += labels.size(0)
            correct_predictions += (predicted == labels).sum().item()

            pred_labels.extend(predicted.cpu().numpy())
            true_labels.extend(labels.cpu().numpy())

            # Store misclassified samples
            for img, true_label, pred_label in zip(images, labels, predicted):
                if true_label != pred_label:
                    misclassified_samples.append((img.cpu(), true_label.cpu(), pred_label.cpu()))

    # Compute accuracy and F1 score
    accuracy = accuracy_score(true_labels, pred_labels) * 100
    f1 = f1_score(true_labels, pred_labels, average='weighted') * 100

    print(f'Accuracy: {accuracy:.2f}%')
    print(f'F1 Score: {f1:.2f}')

    # Plot 6 misclassified samples
    visualize_misclassified_samples(misclassified_samples, class_names)

    return true_labels, pred_labels

def visualize_misclassified_samples(misclassified_samples, class_names, num_samples=6):
    if not misclassified_samples:
        print("No misclassified samples to display.")
        return

    num_samples = min(num_samples, len(misclassified_samples))
    fig, axes = plt.subplots(1, num_samples, figsize=(15, 5))

    for i in range(num_samples):
        img, true_label, pred_label = misclassified_samples[i]
        img = img.numpy()[[3, 2, 1], :, :].transpose((1, 2, 0))  # Select Near-True Color Bands

        axes[i].imshow(img)
        axes[i].axis("off")
        axes[i].set_title(f"True: {class_names[true_label]}\nPred: {class_names[pred_label]}", color="red")

    plt.show()

def plot_confusion_matrix(all_labels, all_predicted):
    confusion_mat = confusion_matrix(all_labels, all_predicted)
    labels = ["0", "1", "2"]
    cm_display = ConfusionMatrixDisplay(confusion_matrix=confusion_mat, display_labels=labels)
    cm_display.plot()
    plt.show()


def tune_hyperparameters(train_dataloader, val_dataloader, test_dataloader, num_epochs):
    learning_rates = [0.1, 0.01, 0.005, 0.001, 0.0005]
    momentums = [0.8, 0.9, 0.99]
    
    best_accuracy = 0
    best_params = {}
    best_model_path = "best_model.pth"

    for alpha in learning_rates:
        for momentum in momentums:
            print(f"Training with alpha={alpha}, momentum={momentum}")
            
            model = create_model()
            train_model(model, train_dataloader, num_epochs, alpha, momentum)
            true_labels, pred_labels = evaluate_model(model, val_dataloader, MODEL_PATH, class_names=["0", "1", "2"])
            plot_confusion_matrix(true_labels, pred_labels)

            accuracy = accuracy_score(true_labels, pred_labels) * 100
            
            if accuracy > best_accuracy:
                best_accuracy = accuracy
                best_params = {"alpha": alpha, "momentum": momentum}
                
                # Save the best model
                torch.save(model.state_dict(), best_model_path)
                print(f"New best model saved with accuracy {accuracy:.2f}% at {best_model_path}")

    print(f"Best hyperparameters: {best_params}, Accuracy: {best_accuracy:.2f}%")

    # Load and evaluate the best model on the test set
    print("\nEvaluating Best Model on Test Set...")
    best_model = create_model()
    best_model.load_state_dict(torch.load(best_model_path))
    evaluate_model(best_model, test_dataloader, best_model_path, class_names=["0", "1", "2"])
````

````python
tune_hyperparameters(train_dataloader, val_dataloader,test_dataloader, num_epochs=EPOCHS_NUM)
````
